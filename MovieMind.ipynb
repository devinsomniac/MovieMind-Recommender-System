```python
m = df['vote_count'].quantile(0.80)
```

```python
C = df['vote_average'].mean()
```

```python
df = df[df['vote_count']>m]
```

```python
def Weight_rating(x,m=m,C=C):
    v = x['vote_count']
    R = x['vote_average']
    WR = (v/(v+m)*R) + (m/(v+m)*C)
    return WR
```

```python
df['score'] = df.apply(Weight_rating,axis=1)
```

```python
df.head()
```

```python
df =df.sort_values(by='score',ascending=False)
```

```python
top_250 = df.head(250)
```

```python
top_250 = top_250.reset_index(drop=True)
```

```python
top_250.head()
```

```python
top_250['genres'] = top_250['genres'].apply(lambda x : [i['name'] for i in x] if isinstance(x,list) else [])
```

```python
top_250.head()
```

```python
top_250['keywords'] = top_250['keywords'].apply(lambda x : [i['name'] for i in x] if isinstance(x,list) else [])
```

```python
top_250.head()
```

```python
top_250.iloc[0]['cast']
```

```python
def clean_cast(x):
    if isinstance(x,list):
        names = [i['name'] for i in x[:5]]
        return names
    return []    
```

```python
top_250['cast'] = top_250['cast'].apply(clean_cast)
```

```python
top_250.iloc[0]['cast']
```

```python
top_250.iloc[0]['crew']
```

```python
def get_director(x):
    if isinstance(x,list):
        for i in x:
            if i['job'] == 'Director':
                director = i['name']
                return director
    return []        
```

```python
top_250['crew'] = top_250['crew'].apply(get_director)
```

```python
top_250.head()
```

```python
def sanitize(x):
    if isinstance(x,list):
       return [str.lower(i.replace(" ","")) for i in x]
    elif isinstance(x,str):
        return str.lower(x.replace(" ",""))
```

```python
features = ['cast','crew','keywords','genres']
for feature in features:
    top_250[feature] = top_250[feature].apply(sanitize)
```

```python
top_250.head(2)
```

```python
top_250.to_csv('Top 250 Pre-processed')
```

```python
top_250['overview'] = top_250['overview'].astype(str)
```

```python
def create_soup(x):
    return ' '.join(x['genres']) + ' ' + x['overview'] + ' ' + ' '.join(x['cast']) + ' ' + x['crew'] + ' ' + ' '.join(x['keywords'])
```

```python
top_250['soup'] = top_250.apply(create_soup,axis=1)
```

```python
top_250.iloc[0]['soup']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vector = CountVectorizer(stop_words = 'english')
```

```python
count_matrix = count_vector.fit_transform(top_250['soup'])
```

```python
from sklearn.metrics.pairwise import linear_kernel
```

```python
cosine_sim = linear_kernel(count_matrix,count_matrix)
```

```python
from fuzzywuzzy import process

def content_based(query, top_250=top_250, cosine_sim=cosine_sim):
    choices = top_250['soup'].tolist()  
    best_match = process.extractOne(query, choices)
    print("Best match:", best_match)
    
    if best_match:
        match_string, score = best_match  
        idx = choices.index(match_string)
        print("Matched index:", idx)
        sim_scores = list(enumerate(cosine_sim[idx]))
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
        sim_scores = sim_scores[1:15]
        movie_indices = [i[0] for i in sim_scores]
        recommended_movies = []
        
        for i in movie_indices:
            recommended_movies.append({
                'title': top_250['title'].iloc[i],
                'poster_path': top_250['poster_path'].iloc[i],
                'genres': top_250['genres'].iloc[i],
                'vote_average': top_250['vote_average'].iloc[i]
            })
        
        return recommended_movies
    
    return "No matches found for your query."
```
